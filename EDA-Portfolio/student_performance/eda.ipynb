## Analysis

### Histograms

```python
sns.histplot(df['math score'], kde=True)
plt.title('Distribution of Math Scores')
plt.show()
```

```python
sns.histplot(df['reading score'], kde=True)
plt.title('Distribution of Reading Scores')
plt.show()
```

```python
sns.histplot(df['writing score'], kde=True)
plt.title('Distribution of Writing Scores')
plt.show()
```

```python
# Average scores by gender
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
```

```python
# Barplot
sns.barplot(x='gender', y='math score', data=df)
plt.title('Average Math Score by Gender')
plt.show()  
sns.barplot(x='gender', y='reading score', data=df)
plt.title('Average Reading Score by Gender')
plt.show()
sns.barplot(x='gender', y='writing score', data=df)
plt.title('Average Writing Score by Gender')
plt.show()
```

```python
# Boxplots
sns.boxplot(x='lunch', y='math score', data=df)
plt.title('Math Score vs Lunch Type')
plt.show()

sns.boxplot(x='lunch', y='reading score', data=df)
plt.title('Reading Score vs Lunch Type')
plt.show()

sns.boxplot(x='lunch', y='writing score', data=df)
plt.title('Writing Score vs Lunch Type')
plt.show()
```

```python
# Compare scores with/without test prep
df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
```

```python
# Correlation heatmap
sns.heatmap(df[['math score', 'reading score', 'writing score']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Scores')
plt.show()
```

### Key Findings from EDA on Students Performance Dataset

- Female students outperform male students in reading and writing on average.
- Male students slightly outperform females in math on average.
- Students who completed the test preparation course have higher average scores across all subjects.
- Standard lunch is associated with higher performance compared to free/reduced lunch.
- There is a strong positive correlation between reading and writing scores.
- Parental level of education has a subtle impact—students with more educated parents tend to perform slightly better.
